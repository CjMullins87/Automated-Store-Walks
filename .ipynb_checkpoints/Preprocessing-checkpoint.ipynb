# Preprocessing: Making Sure our Data is Read Correctly

We have a lot of unique features in our dataset that come from the way BI handles data. BI handles negatives like a ledger, so rather than including negative values, it presents them in parentheseses, e.g. -3 becomes (3); -16 becomes (16). If you're familiar with this system, it's fairy intuitive, but an issue that arises here is that these numbers in parentheses don't get handled as negatives. They get handled as objects.

The same issue arises in large numbers - BI will include commas so that they're easier for us to read. However, this makes it less easy for Pandas to read, which then interprets the number as an object, too. 

I found these issues while trying to do quick conversion on columb types in our dataset, as shown below.

```python
BVS26wk['Bill Units'].astype(np.int64)
```

All the way down at the bottom we can see it's having a hard time processing the value "1,008". We can hunt down problem values in this column quickly by using a try/except system, as shown below:

```python
for i in BVS26wk['Bill Units']:
    try:
        int(i)
    except:
        print(i)
```

```python
for i in BVS26wk['Est. Units On Hand']:
    try:
        int(i)
    except:
        print(i)
```

The issue here is just the way that BI presents data. We're going to throw together a couple helper functions to clean this text data up into a usable format - helper functions are espescially important since we'll be doing this across multiple datasets.

```python
def negatives(value):         
    try:    
        neg = value.replace('(', '-')
        neg = neg.replace(')', '')
    except:
        return int(value)
    return int(neg)

def big_numbers(value):
    try:    
        num = value.replace(',', '')
    except:
        return int(value)
    return int(num)

def dollars(value):
    numA = value.replace('$', '')
    try:
        numB = big_numbers(numA)
    except:
        return negatives(numA)
    return int(numB)
```

These should help us handle some of our unique cases, but I'll need something to handle the information in our estimated sell through percentage.

```python
BVS26wk['Est. Sell Thru % (Units)'].isnull().sum()
```

Right now in our big set there's only 10 units where the sell through percentage is negative. Which is good - we want to be selling things that we're going to order and sell more often than just once in a timespan > 6months.

```python
def percent(value):
    try:
        corr = int(value.replace('%', ''))
    except:
        corr = int(-1)
    return(corr)
```

```python
a = np.nan
percent(a)
```

```python
percent('56%')
```

Okay, looks like it's up and running. Let's apply our functions.

```python
datasets = [BVS26wk, BVS12wk, BVS7]

for data in datasets:
    data['Est. Sell Thru % (Units)'] = data['Est. Sell Thru % (Units)'].apply(lambda x: percent(x))
    data['Est. Units On Hand'] = data['Est. Units On Hand'].apply(lambda x: negatives(x))
    data['Bill Units'] = data['Bill Units'].apply(lambda x: big_numbers(x))
    data['Scan Units'] = data['Scan Units'].apply(lambda x: big_numbers(x))
    data['Bill $'] = data['Bill $'].apply(lambda x: dollars(x))
    data['Sales'] = data['Sales'].apply(lambda x: dollars(x))
```

```python
copy = BVS7DOS

copy['Est. Sell Thru % (Units)'] = copy['Est. Sell Thru % (Units)'].apply(lambda x: percent(x))
copy['Est. Units On Hand'] = copy['Est. Units On Hand'].apply(lambda x: negatives(x))
copy['Bill Units'] = copy['Bill Units'].apply(lambda x: big_numbers(x))
copy['Scan Units'] = copy['Scan Units'].apply(lambda x: negatives(x))
copy['Bill $'] = copy['Bill $'].apply(lambda x: dollars(x))
copy['Sales'] = copy['Sales'].apply(lambda x: dollars(x))
```
